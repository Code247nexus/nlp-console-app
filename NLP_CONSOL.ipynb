```python
# creating a nlp app
# import nlpcloud
import nlpcloud

class NlpApp:
  def __init__(self):
    self.__database = {}
    self.first_menu()


  def first_menu(self):
    user1_input = input("""
    HII THERE! HOW WOULD YOU LIKE TO PROCEED ?
    1)NEW HERE ? REGISTER
    2)ALREADY REGISTERED ?LOGIN
    3)EXIT

    """)

    if user1_input == '1':
      self.__register()

    elif user1_input == '2':
      self.__login()

    else:
      exit()


  def __register(self):

    name = input("enter your name: ")
    email = input("enter you email: ")
    password = input("enter you password: ")
    if email in self.__database:
      print("email already registered")
      print("proceed to login")
      self.__login()

    else:
      self.__database[email]=[name,password]
      print(" registered successfully.kindly move to login")
      self.first_menu()

  def __login(self):
    print("welcome to login interface: ")
    email = input("enter you email:")
    password = input("enter your password:")

    if email in self.__database:
      if self.__database[email][1] == password:
        print("login successful")
        self.second_menu()
      else:
        print(" your password is wrong !")
        self.first_menu()
    else:
      print("your email is not registered kindly register first")
      self.first_menu()

  def second_menu(self):
    user2_input = (input("""
  WELCOME TO NLP APLLICATION
  HOW WOULD YOU LIKE TO PROCEED?
  1)CHATBOT/CONVERSATIONAL AI
  2)CODE GENERATION
  3)SUMMARIZATION
  4)TRANSLATION
  5)GRAMMER AND SPELLING CORRECTION
  """))

    if user2_input == '1':
      self.chat_bot()
    elif user2_input == '2':
      self.code_generation()

    elif user2_input == '3':
      self.summarization()

    elif user2_input == '4':
      self.translation()

    elif user2_input == '5':
      self.grammer()

    else:
      exit()


  def chat_bot(self):
    u_input = str(input("""what would you like to talk about: """))
    client = nlpcloud.Client("gpt-oss-120b", "YOUR API KEY HERE", gpu=True)
    response = client.chatbot(u_input,context="""This is a discussion between a human and an AI.The AI is called Patrick.""",history=[ ])
    print(response)
    u2_input = input("do you want to continue the chat:(1 - true /0 - false)")
    if u2_input == '1':
      self.chat_bot()
    else:
      print(" thank you ")
      self.second_menu()

  def code_generation(self):
    code_brief = input(""" what you wolud like to code about plovide statement of question: """)
    client = nlpcloud.Client("gpt-oss-120b", "YOUR API KEY HERE", gpu=True)
    response = client.code_generation(code_brief)
    print(response['generated_code'])
    self.second_menu()

  def summarization(self):
    para = input("enter your paragraph: ")
    client = nlpcloud.Client("gpt-oss-120b", "YOUR API KEY HERE", gpu=True)
    response = client.summarization(para,size="large")
    print(response['summary_text'])
    self.second_menu()

  def translation(self):
    statement_to_convert = input("enter the text to convert")
    # target = input(""" enter your target lnguage :
    # select 1)bho_Deva -> BHOJPURI
    #        2)fra_Latin ->FRENCH
    #        3)"ace_Arab" ->ARABIAN
    #        4)"eng_Latn"->ENGLISH
    # """)
    # target_input = " "
    # if target =='1':
    #   target_input = "bho_Deva"

    # elif target == '2':
    #   target_input ="fra_Latin"
    # elif target == '3':
    #   target_input = "ace_Arab"
    # else:
    #     target_input = "eng_Latn"

    client = nlpcloud.Client("nllb-200-3-3b", "YOUR API KEY HERE", gpu=False)
    response =client.translation(statement_to_convert,source="eng_Latn",
    target="ace_Arab")
    print(response['translation_text'])
    self.second_menu()

  def grammer(self):
    statement = input(""" enter your text for correction: """)
    client = nlpcloud.Client("gpt-oss-120b", "YOUR API KEY HERE", gpu=True)
    response = client.gs_correction(statement)
    print(response)
    self.second_menu()
```
